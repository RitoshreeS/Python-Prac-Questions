1. <font color="blue"> **For every Indian state, considering a particular month (Last2Digits_Of_RollNumber%10 + 3), find the maximum cases reported for confirmed, deaths, and recovered cases  along with date on which these cases were reported in year 2020. Display the result in the self-explanatory format. (PS: Consider month number 1 denotes January)**</font>

```python
### TO DO ###
#May=(62%10+3=5 month)
df['Month'] = pd.DatetimeIndex(df['Date'],dayfirst=True).month
df['Year'] = pd.DatetimeIndex(df['Date'],dayfirst=True).year
df.query("Month == 5 and Year == 2020").groupby(["State/UnionTerritory","Month","Date"]).max()
```

2. <font color="blue"> **For the month considered in previous case, identify the number of cases on the 6th day of the month.**</font>
---



```python
### TO DO ###
df['Day'] = pd.DatetimeIndex(df['Date'],dayfirst=True).day
df.query("Day==6 and Month == 5").groupby(["State/UnionTerritory","Month","Confirmed","Day"]).max()
```
