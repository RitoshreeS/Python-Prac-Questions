```python
#a. Compute the mean, standard deviation, and variance of a two dimensional random integer array along the second axis.
import numpy as np
A=np.arange(1,21).reshape(4,5)
x=np.mean(A)
print("Mean: ",x)
y=np.std(A)
print("Standard Deviation: ",y)
z=np.var(A)
print("Variance: ",z)
```

```python
#b. Get the indices of the sorted elements of a given array.
#   B = [56, 48, 22, 41, 78, 91, 24, 46, 8, 33]
import numpy as np
B=np.array([56, 48, 22, 41, 78, 91, 24, 46, 8, 33])
print("Original array:",B)
A=np.argsort(B)
print("Indices of the sorted elements of original array:")
print(A)
```

```python
#c. Create a 2-dimensional array of size m x n integer elements, also print the shape, type and data type of the array
#   and then reshape it into nx m array, n and m are user inputs given at the run time.
import numpy as np
x=eval(input("Enter the no. of rows:"))
y=eval(input("Enter the no. of columns:"))

A=np.arange(1,31).reshape(x,y)
print("Matrix A:")
print(A)
print("Shape of array:",A.shape)
print("Data type of array:",A.dtype)

A=np.arange(1,31).reshape(y,x)
print("After reshaping:")
print(A)
```

```python
#Q1 Write a program to create an NumPy array of all the even integers from 30 to 70.
import numpy as np
A=np.arange(30,71,2)
print("Even integers from 30 to 70:",A)
```

```python
#Q2 Write a program to create a 10x10 NumPy matrix, in which the elements on the borders will be equal to 1, and inside 0.
import numpy as np
A= np.ones((10, 10))
A[1:-1, 1:-1] = 0
print("Matrix:")
print(A)
```

```python
#Q3 Write a program to create a 5x10 NumPy matrix of distinct numbers from 1 to 50 and swap 2nd and 3rd column in a 2d NumPy array.
import numpy as np
A=np.arange(1,51).reshape(5,10)
print("Original Matrix:")
print(A)
A[:,[1,2]]=A[:,[2,1]]
print("Modified Matrix:")
print(A)
```

```python
#Q4 Write a program to create a 5x10 NumPy matrix of distinct numbers from 1 to 50 and swap 2nd and 3rd row in a 2d NumPy array.

import numpy as np
A=np.arange(1,51).reshape(5,10)
print("Original Matrix:")
print(A)
A[[1,2]]=A[[2,1]]
print("Modified Matrix:")
print(A)

```

```python
#Q5 Add the two 2d NumPy arrays and modify the resulting array by calculating the square root of each element
import numpy as np
A=np.arange(1,21).reshape(4,5)
B=np.arange(21,41).reshape(4,5)
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
C=np.add(A,B)
print("Sum:")
print(C)
D=np.sqrt(C)
print("Modified matrix:")
print(D)
```

```python
#Q6 Create a 5X2 NumPy integer array from a range between 100 to 200 such that the difference between each element is 10.
#   Retrive indexes of only those elements where numbers are divisible by 3.

import numpy as np
A=np.arange(100,200,10).reshape(5,2)
print("Matrix A:")
print(A)
print("Elements divisible by 3:",np.argwhere(A%3==0))
```
